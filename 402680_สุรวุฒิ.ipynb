```python
def clean():
  for i in range (0, len(data['_id']), 1):
    cleana = str(data['co'][i])
    data['co'][i] = re.sub(r",+", ",", cleana) #ลบ ,, ให้เป็น ,
    data['co'][i] = re.sub(r"\*", "", cleana) #สบคำที่เป็นดอกจันทร์


clean()
```

```python
# ลบช่องว่างที่มากกว่า 1
def clean1():
  for i in range (0, len(data['_id']), 1):
    clean = str(data['co'][i])
    data['co'][i] = clean.replace('  ', ' ')

clean1()
```

```python
#ตรวจสอบว่าใน 'co' มีคำที่ค้นหาไหม
def check(search):
  n = 0
  for i in range (0, len(data['_id']), 1):
    if search in str(data['co'][i]):
      n = n + 1
      print(i, data['_id'][i], data['co'][i])
  print('มีจำนวน', n, 'แถว')
```

```python
# ลบคำจำพวก นาย, นางสาว, นาง
def delete_artical():
  for i in range (0, len(data['_id']), 1):
    text = str(data['co'][i])
    text = text.split(', ')
    for j in range (0, len(text), 1):
      if re.match((r"^นาย"), text[j]):
        text[j] = re.sub((r"^นาย"), '', text[j])
      elif re.match((r"^นางสาว"), text[j]):
        text[j] = re.sub((r"^นางสาว"), '', text[j])
      elif re.match((r"^นาง"), text[j]):
        text[j] = re.sub((r"^นาง"), '', text[j])
    text = ', '.join(text)
    data['co'][i] = text

delete_artical()
```

```python
# ลบคำทั้งคำที่มีคำนั้นปรากฎ
def delete_whole(letter):
  for i in range (0, len(data['_id']), 1):
    delete = str(data['co'][i])
    if letter in delete:
      list_delete = delete.split()
      j = 0
      while j < len(list_delete):
        if letter in list_delete[j]:
          list_delete.remove(list_delete[j])
        j = j + 1
      delete = ' '.join(list_delete)
      data['co'][i] = delete
```

```python
# ลบเฉพาคำนั้นๆ
def delete_point(point):
  for i in range (0, len(data['_id']), 1):
    point_text = str(data['co'][i])
    if point in point_text:
      point_text = point_text.replace(point, '')
    data['co'][i] = point_text
```

```python
def capital():
  for i in range (0, len(data['id']), 1):
    text = data['name'][i]
    list_text = text.split(' ')
    list_n = []
    for j in list_text:
      string = j.capitalize()
      list_n.append(string)
      text_n = ' '.join(list_n)
    data['name'][i] = text_n
```
